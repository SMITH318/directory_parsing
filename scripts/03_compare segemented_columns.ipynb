```python
import json
import os
from pathlib import Path
```

```python
project_root = Path(os.getcwd()).parent

input_dir = project_root / "data" / "03_processed"
input_file = input_dir / "articles.json"
with open(input_file, 'r', encoding='utf-8') as f:
    all_metadata = json.load(f)
```

```python
articles = all_metadata['articles']
amd1918_p571s = [a for a in articles if a['article_id'].startswith('AMD1918_p0571')]
len(amd1918_p571s)
```

```python
source_ids = set([a["source_pdf"] for a in amd1918_p571s])
source_ids
```

```python
for source_id in source_ids:
    print(f"Source ID: {source_id}")
    with open(input_dir / f"{source_id}.txt", 'w', encoding='utf-8') as f:
        page_articles = [a for a in amd1918_p571s if a['source_pdf'] == source_id]
        for col in range(3):
            col_articles = [a for a in page_articles if int(a.get('column')) == col]
            f.write(f"=== Column {col} ===\n")
            for article in col_articles:
                f.write(article['full_text'] + "\n\n")
```
